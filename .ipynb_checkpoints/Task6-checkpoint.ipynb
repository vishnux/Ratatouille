```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained('TrainedModels')
model = AutoModelForCausalLM.from_pretrained('TrainedModels')
```

```python
import pandas as pd
df = pd.read_pickle('df_recipes_processed.pkl')
```

```python
df[df.combined.str.contains('Use a slotted spoon to remove the ')].combined.tolist()
```

```python
with open('dataset_val.txt', 'r') as f:
    data = f.read()
```

```python
special_token = '<|endoftext|>'
data = data.split('<|endoftext|>')
```

```python
sample = data[0].split('Instructions:')[0] + 'Instructions:'
print(sample)
```

```python
%%time
prompt_text = sample

# prompt_text = 'Ingredients: 5 eggs \n 10 grams butter \n breadcrumbs \n cheese \n Instructions:'
encoded_prompt = tokenizer.encode(prompt_text, add_special_tokens=False, return_tensors="pt")

input_ids = encoded_prompt

output_sequences = model.generate(
    input_ids=input_ids,
    max_length=len(encoded_prompt[0])+800,
    temperature=0.9,
    top_k=100,
    top_p=0.9,
    repetition_penalty=1,
    do_sample=True,
    num_return_sequences=5,
)
```

```python
for i, output_sequence in enumerate(output_sequences):
    result = tokenizer.decode(output_sequence)
    
    result = result[:result.index(special_token)]
    print('--------------------')
    print(f'Generated Output #{i+1}')
    print('--------------------')
    print(result)
```
