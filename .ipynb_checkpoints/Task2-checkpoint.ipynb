# Using Natural Language Processing to Generate New Recipes in Python
## Task 2: Exploratory Data Analysis and Preprocessing

```python
import json
from collections import defaultdict

import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#matplotlib.rcParams['figure.dpi'] = 800
import matplotlib.pyplot as plt
plt.style.use('ggplot')

import pandas as pd
from tqdm.notebook import tqdm

import numpy as np

import seaborn as sns
```

```python
data = json.load(open('Dataset/recipes_raw_nosource_ar.json', 'r'))
```

```python
keys = list(data.keys())
data[keys[0]]
```

```python
recipe_sources = ['ar', 'epi', 'fn']
```

```python
df = pd.DataFrame()
sources, titles, ingredients, instructions = [], [], [], []
for recipe_source in recipe_sources:
    data = json.load(open(f'Dataset/recipes_raw_nosource_{recipe_source}.json', 'r'))
    for _, recipe in data.items():
        if ('title' in recipe) and ('ingredients' in recipe) and ('instructions' in recipe):
            sources.append(recipe_source)
            titles.append(recipe['title'])
            ingredients.append([str(ingredient).replace('ADVERTISEMENT', '') for ingredient in recipe['ingredients']])
            instructions.append(str(recipe['instructions']).replace('ADVERTISEMENT', '').replace('\n', ' '))
            
df['source'] = sources
df['title'] = titles
df['ingredients'] = ingredients
df['instructions'] = instructions
```

```python
df.head(2)
```

```python
df['ingredient_count'] = df['ingredients'].str.len().fillna(0).astype(int)
df['instruction_length'] = df['instructions'].str.split().str.len().fillna(0).astype(int)
```

```python
df.to_pickle('df_recipes_raw.pkl')
```

```python
sns.boxplot(data=df, x='source', y='ingredient_count', showfliers=False)
```

```python
sns.boxplot(data=df, x='source', y='instruction_length', showfliers=False)
```

# Playground

```python
stats = defaultdict()
for source, recipes in data.items():
    temp = defaultdict(list)
    for key, recipe in recipes.items():
        if ('ingredients' in recipe) and ('instructions' in recipe):
            temp['ingredient_item_counts'].append(len(recipe['ingredients']))
            temp['instructions_word_counts'].append(len(str(recipe['instructions']).split()))
    stats[source] = temp
```

```python
df = pd.DataFrame()
```

```python
sources, ingredient_item_counts, instructions_word_counts = [], [], []
for k, v in stats.items():
    sources.extend([k]*len(v['ingredient_item_counts']))
    ingredient_item_counts.extend(v['ingredient_item_counts'])
    instructions_word_counts.extend(v['instructions_word_counts'])

df['source'] = sources
df['ingredient_item_count'] = ingredient_item_counts
df['instructions_word_count'] = instructions_word_counts
```

```python
sns.boxplot(data=df, x='source', y='ingredient_item_count', showfliers=False)
```

```python
sns.boxplot(data=df, x='source', y='instructions_word_count', showfliers=False)
```
